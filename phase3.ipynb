```python
sequencesData = np.zeros((24,3))
deltaData = np.zeros((24,3)) 
rotationData  = np.zeros((24,3,3)) 
for i in range(24):
    print(i)
    squence = sequences[i]
    d1,d2,d3 =  actSequence(squence,10,3,10000)
    sequencesData[i] = d1
    deltaData[i] = d2
    rotationData[i] = d3
```

```python
rotorStep = 50_000
rotorR = np.array([0.,0.,0.])
rotorRotation = np.eye(3)
p = 0.001
s = 1

poses = np.zeros((rotorStep,3))
for i in range(rotorStep):
    if np.random.uniform(0,1) <= p:
        s = np.random.randint(0,24)
    rotorR += np.matmul(rotorRotation,deltaData[s])
    rotorRotation = np.matmul(rotationData[s],rotorRotation)
    poses[i,:] = rotorR


ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')

ax.scatter(xs = poses[:,0],ys = poses[:,1], zs=poses[:,2],s=1)

plt.savefig("random_walk_p=0.001.png")
    
        
```

```python
def reciprocity(alpha,p):
    return (p*np.exp(alpha))/(1-p+p*np.exp(alpha))

```

```python
def conncetance(alpha,p,n):
    A  = 2*(p*(np.exp(alpha) -1)*(1-p)**2 + (1-p)*(1+(np.exp(alpha)-1)*p**2))/(1-p)**4
    return p*(1-p)* (n*(n-1)/2)* ((1-p)**2 / (1+ (np.exp(alpha)-1)*p**2)) *A /(n**2)
```

```python
import networkx as nx
import numpy as np
from sklearn.metrics.cluster import normalized_mutual_info_score

# Function to create a 2-layer graph
def create_graph(n, p):
    G = nx.Graph()
    for i in range(2):
        nodes = range(i * n, (i + 1) * n)
        G.add_nodes_from(nodes, layer=i)
        G.add_edges_from(nx.fast_gnp_random_graph(n, p).edges(), layer=i)
    return G


# Function to calculate NMI
def calculate_nmi(G):
    layer0 = [n for n, d in G.nodes(data=True) if d['layer']==0]
    layer1 = [n for n, d in G.nodes(data=True) if d['layer']==1]
    return layer0,layer1,normalized_mutual_info_score(layer0, layer1)

# Create two 2-layer graphs
G1 = create_graph(10, 0.5)
G2 = create_graph(10, 1)

# Calculate NMI for the two graphs
nmi_G1 = calculate_nmi(G1)
nmi_G2 = calculate_nmi(G2)

print(f"NMI for G1: {nmi_G1}")
print(f"NMI for G2: {nmi_G2}")
```

```python


@njit
def getPs(sigmas,k,mu,q):
    ps = np.zeros((sigmas.shape[0],sigmas.shape[0]))
    for i in range(sigmas.shape[0]):
        for j in range(sigmas.shape[0]):
            if sigmas[i]==sigmas[j]:
                ps[i,j] = k*(1-mu)/(q-1)
            else:
                ps[i,j] = k*mu/(n-q)
    return ps

@njit
def getA(sigmas,k,mu,q):
    ps = getPs(sigmas,k,mu,q)
    rands = np.random.random((sigmas.shape[0],sigmas.shape[0]))
    return rands <= ps
```

```python
def create_graph_from_adjacency_matrices(adj_matrix1, adj_matrix2):
    G = nx.Graph()
    for i, adj_matrix in enumerate([adj_matrix1, adj_matrix2]):
        nodes = range(i * len(adj_matrix), (i + 1) * len(adj_matrix))
        G.add_nodes_from(nodes, layer=i)
        G.add_edges_from([(i+n, j+n) for i, row in enumerate(adj_matrix) for j, value in enumerate(row) if value], layer=i)
    return G
```

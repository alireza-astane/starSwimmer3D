```julia
img = CairoRGBSurface(512, 128);
img = CairoRGBSurface(512, 128);
cr = CairoContext(img);
save(cr);
```

```julia
set_source_rgb(cr, 0.5, 0.5, 0.5);

rectangle(cr, 0.0, 0.0, 512.0, 128.0);

fill(cr);

restore(cr);

save(cr);
```

```julia
x0=61.2; y0=74.0;

x1=214.8; y1=125.4;

x2=317.2; y2=22.8;

x3=470.8; y3=74.0;

move_to(cr, x0, y0);

curve_to(cr, x1, y1, x2, y2, x3, y3);

set_line_width(cr, 10.0);

stroke_preserve(cr);

restore(cr);
```

```julia
move_to(cr, 12.0, 12.0);

set_source_rgb(cr, 0, 0, 0);

show_text(cr,"Line_Figure")

write_to_png(cr,"Line_Figure.png");
```

```julia
x = range(0, stop=3pi, length=100);

c = cos.(x);

s = sin.(x);
p = FramedPlot(
                        title="Winston Graphics!",
                        xlabel="\\Sigma x^2_i",
                        ylabel="\\Theta_i")
                        
add(p, FillBetween(x, c, x, s))

add(p, Curve(x, c, color="black"))

add(p, Curve(x, s, color="red"))
```
